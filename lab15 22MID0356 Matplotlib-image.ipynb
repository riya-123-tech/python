```python
from PIL import Image

import matplotlib.pyplot as plt
import numpy as np
```

```python
# Opens the ladybug image from Downloads, converts it to a NumPy array, and prints its data. 
img = np.asarray(Image.open('/Users/riyaparikh/Downloads/ladybug.jpeg'))
print(repr(img))
```

```python
 #Display the NumPy image array as a visual image using Matplotlib.
imgplot = plt.imshow(img)
```

```python
# Extracts the red channel (first layer) from the image and displays it as a grayscale image. 
lum_img = img[:, :, 0]
plt.imshow(lum_img)
```

```python
# Displays the red channel of the image using the "hot" colormap for a heatmap effect.
plt.imshow(lum_img, cmap="hot")
```

```python
# Shows the red channel of the image and changes its colormap to 'nipy_spectral' for visualization. 
imgplot = plt.imshow(lum_img)
imgplot.set_cmap('nipy_spectral')
```

```python
# Displays the red channel of the image and adds a color scale bar to indicate intensity values. 
imgplot = plt.imshow(lum_img)
plt.colorbar()
```

```python
# Flattens the red channel to 1D and plots its pixel intensity distribution as a histogram with black bars. 
plt.hist(lum_img.ravel(), bins=range(256), fc='k', ec='k')
```

```python
imgplot = plt.imshow(lum_img)  # Display the red channel of the image
imgplot.set_clim(0, 175)       # Set the color scale to map intensities from 0 to 175
plt.colorbar()                  # Optional: add a colorbar to see the intensity mapping
plt.show()
```

```python
#REMOVE BACKGROUND
from PIL import Image
import numpy as np

# Open the image
img = Image.open('/Users/riyaparikh/Downloads/ladybug.jpeg').convert('RGBA')  # ensure it has alpha channel

# Convert to NumPy array
data = np.array(img)

# Define background color to remove (here, white)
r, g, b, a = data[:,:,0], data[:,:,1], data[:,:,2], data[:,:,3]
mask = (r > 240) & (g > 240) & (b > 240)  # detect white pixels
data[mask] = [255, 255, 255, 0]  # set those pixels fully transparent

# Convert back to image
new= Image.fromarray(data)

# Save or show
new.save('/Users/riyaparikh/Downloads/ladybug_no_bg.png')
new.show()
```

```python
plt.imshow(new)
```

```python

# Load the background-removed image
new_img = Image.open('/Users/riyaparikh/Downloads/ladybug_no_bg.png')
data = np.array(new_img)

# Extract red channel and alpha channel
lum_new = data[:, :, 0]  # red channel
alpha = data[:, :, 3]    # alpha channel

# Plot histogram ignoring transparent pixels
plt.hist(lum_new[alpha > 0].ravel(), bins=range(256), fc='k', ec='k')
plt.title('Red Channel Histogram (Background Removed)')
plt.xlabel('Pixel Intensity (0-255)')
plt.ylabel('Frequency')
plt.show()
```
